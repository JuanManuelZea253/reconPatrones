Estructurando datos en Python

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/PatronesClass2022/Clase3/soundImages/"
```

```python
def dataCleaner (extensions, files):
  for file_ in os.listdir(path):

    if os.path.splitext(file_)[1].lower() in extensions:

      files.append(path + file_)

return files
```

```python
extensions = ['.jpeg','.png','.jpg','.tif','.bmp','.gif']
files = []
dataCleaner(extensions, files)
```

```python
print (len(files))
```

```python
images = []
for image in files:
  image = cv2.imread(image)
  # -- preprocesamiento siempre antes de que se carguen las imagenes -- #
  # reescalado, filtrado y cualquier preprocesamiento que sea siempre antes de cargar
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  image = cv2.resize(image,(256,256))
  images.append(image)

print (len(images))
```

```python
#pone la imagen en columnas formando una figura(_,) y sus plots
_,axes =  plt.subplots(1,len(images),figsize = (30,8))
axes =  axes.ravel()

for i, image in enumerate(images): 
  axes[i].imshow(image)
  #plt.show
```

Estructurar audios

```python
import librosa
from librosa.display import waveplot 
```

```python
audioExtensions =  ['.mp3', '.mp4', '.mid', '.midi', '.au', '.aiff', '.wav', '.wma' ]
audioFiles = []

dataCleaner (audioExtensions, audioFiles)
print(audioFiles)
print(len(audioFiles))
```

```python
audioSignals = []
for signal in audioFiles:

  audio, muestringFrecuense = librosa.load(signal)
  audioSignals.append([audio,muestringFrecuense])

print(len(audioSignals))
```

```python
for i,soundRate in audioSignals:
  waveplot(i,soundRate)
  plt.figure()
  plt.title('Audio Loaded')
```
