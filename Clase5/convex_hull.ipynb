```python
import matplotlib.pyplot as plt

from skimage.morphology import convex_hull_image
from skimage import img_as_float
import cv2
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

# Load image - horse.jpg (grayscale)
image = cv2.imread('/content/drive/MyDrive/PatronesClass2022/Repositorio/patrones/Clase5/horse.jpg',0)

# Binarize image (as required by convex_hull_image)
image = image > 127

c_hull = convex_hull_image(image)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Display original image
axes[0].set_title('Original picture')
axes[0].imshow( image , cmap = "gray")
axes[0].set_axis_off()

# Display convex hull transform
axes[1].set_title('Transformed picture')
axes[1].imshow(c_hull, cmap = "gray")
axes[1].set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# Estimate the difference between convex hull and original image
c_hull_diff = img_as_float( c_hull )
print (c_hull_diff.max(), c_hull_diff.min())

# use other form
c_hull_diff = c_hull * 1.0
print (np.max(c_hull_diff))
c_hull_diff[image] = 2

fig, ax = plt.subplots()
ax.imshow(c_hull_diff, cmap = "gray")
ax.set_title('Difference')
ax.set_axis_off()
plt.show()
```
