```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure
```

## **Using a random surface**

```python
# Construct some test data
x, y = np.ogrid[-np.pi:np.pi:100j, -np.pi:np.pi:100j]
r = np.sin(np.exp((np.sin(x)**3 + np.cos(y)**2)))

plt.imshow(r, cmap = 'gray')
```

```python


# Find contours at a constant value of 0.8
contours = measure.find_contours( r , 0.1)#nivel sensibilidad de la que se quiere hacer la detecci√≥n de contornos // cambios abruptos (0) no tanto en adelante hasta (1)

# Display the image and plot all contours found
fig, ax = plt.subplots(figsize = (10,10))
ax.imshow( r , cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=3)#array de listas n es el indice y se lee en filas y columnas que se presentan en la funcion
    #cada color es un contorno distinto

ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
