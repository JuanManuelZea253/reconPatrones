```
# Importar librerias. Incluya acá todas las librerías que requiera
import cv2
import numpy as np
from skimage import measure
from skimage.measure import find_contours, approximate_polygon, subdivide_polygon
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```
#---- Introduzca su codigo aqui. Puede utilizar cuantas celdas de código considere necesarias
img_c = cv2.imread('/content/drive/MyDrive/PatronesClass2022/Repositorio/patrones/entregas/figures_1.jpg',0)
img = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)
imgGray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
_,threshold = cv2.threshold( imgGray, 242, 255, cv2.THRESH_BINARY)
contours = find_contours(threshold, 0)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,6))
ax1.imshow(img)
ax2.imshow(threshold)

triangulos = 0
rectangulos = 0
pentagonos = 0
heptagonos = 0
hexagonos = 0
octagonos = 0 
circulos = 0



for n, contour in enumerate(contours):
    coords = approximate_polygon(contour, tolerance=15)

    if len(coords)-1 == 3:
      triangulos = triangulos + 1
    elif len(coords)-1 == 4:
      rectangulos = rectangulos + 1
    elif len(coords)-1 == 5:
      pentagonos = pentagonos +1
    elif len(coords)-1 == 6:
      hexagonos = hexagonos + 1
    elif len(coords)-1 == 7:
      heptagonos = heptagonos + 1
    elif len(coords)-1 == 8:
      octagonos = octagonos + 1
    elif len(coords)-1 > 8 :
      circulos = circulos + 1

    ax3.plot(coords[:, 1], coords[:, 0], linewidth=2)

```

```
print('triangulos =' + str(triangulos) +'\n' 
      'rectangulos = ' + str(rectangulos) + '\n'
      'pentagonos = ' + str(pentagonos) + '\n'
      'hexagonos = ' + str(hexagonos) + '\n'
      'heptagonos = ' + str(heptagonos) + '\n'
      'octagonos = ' + str(octagonos) + '\n'
      'circulos = ' + str(circulos) + '\n'
      )

#---- Imprima la cantidad aproximada de cada tipo de figura
```
