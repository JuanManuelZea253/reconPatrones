## **Ejercicio 6**

Cree una función que tome el nombre de un archivo con su extensión en formato string y retorne una tupla en el qué, la primera posición sea el nombre del archivo y la segunda la extensión. \\
**NOTA:** NO use splitext ni similares.

```
# 
def takePathFile(archive):
  if archive.count('.')>1:
    file_ = archive
  else:
     check = archive.find('.')
     extension = archive[check:]
     nombre = archive[:check]
     file_ = (nombre,extension)

  return file_
print(takePathFile('/content/drive/MyDrive/PatronesClass2022/lena.jpg'))

```
